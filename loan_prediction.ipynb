```python
train = pd.read_csv("train_dataset.csv")
test = pd.read_csv("test_dataset.csv")
```

```python
train = train.drop("Loan_ID", axis=1)
test  = test.drop("Loan_ID", axis=1)
```

```python
mode_cols = ['Gender','Married','Dependents','Self_Employed','Education','Property_Area']
median_cols = ['LoanAmount','Loan_Amount_Term']
mode_credit = ['Credit_History']
```

```python
for col in mode_cols:
    train[col].fillna(train[col].mode()[0], inplace=True)
    test[col].fillna(test[col].mode()[0], inplace=True)
```

```python
for col in mode_credit:
    train[col].fillna(train[col].mode()[0], inplace=True)
    test[col].fillna(test[col].mode()[0], inplace=True)
```

```python
for col in median_cols:
    train[col].fillna(train[col].median(), inplace=True)
    test[col].fillna(test[col].median(), inplace=True)
```

```python
train.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

for col in train.columns:
    if train[col].dtype == "object":
        train[col] = encoder.fit_transform(train[col])

for col in test.columns:
    if test[col].dtype == "object":
        test[col] = encoder.fit_transform(test[col])
```

```python
X = train.drop("Loan_Status", axis=1)
y = train["Loan_Status"]
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
```

```python
x_ros,y_ros = ros.fit_resample(X,y)
```

```python
y_ros.value_counts()
```

```python
from sklearn.model_selection import train_test_split



X_train, X_test, y_train, y_test = train_test_split(x_ros, y_ros, test_size=0.2, random_state=42)
```

### Auto ML

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
dtc = DecisionTreeClassifier()
rfc = RandomForestClassifier()
abc = AdaBoostClassifier()
gbc = GradientBoostingClassifier()
log = LogisticRegression()
knc = KNeighborsClassifier()
svc = SVC()
```

```python
models = [dtc, rfc, abc, gbc, log, knc, svc]
models
```

### Auto ML with Normal Cross Validation

```python
for m in models:
    scores = cross_val_score(m, x_ros, y_ros, cv = 20, n_jobs = -1)
    scores.mean()
    print(m, "==>", scores.mean())
```

### Auto ML with Stratified KFold Cross Validation

```python
for m in models:
    skf = StratifiedKFold(n_splits = 20)
    scores = cross_val_score(m, x_ros, y_ros, cv = skf, n_jobs = -1)
    scores.mean()
    print(m, "==>", scores.mean())
```
